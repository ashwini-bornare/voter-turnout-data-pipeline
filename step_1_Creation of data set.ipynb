### Importing necessary packages

```python
import os
import pandas as pd
```

### Reading input data

```python
voters_df_2014 = pd.read_excel('pc_wise_voters_turn_out_2014.xlsx')
voters_df_2019 = pd.read_excel('pc_wise_voters_turn_out_2019.xls')
voters_df_2024 = pd.read_excel('pc_wise_voters_turn_out_2024.xls')
```

### String the left and right spaces from all the string columns in the dataframe


```python
for col in voters_df_2014.select_dtypes(include=['object']).columns:
    voters_df_2014[col] = voters_df_2014[col].str.strip()
for col in voters_df_2019.select_dtypes(include=['object']).columns:
    voters_df_2019[col] = voters_df_2019[col].str.strip()
for col in voters_df_2024.select_dtypes(include=['object']).columns:
    voters_df_2024[col] = voters_df_2024[col].str.strip()
```

### Verifying that all files have same structure, columns etc.


```python
print(voters_df_2014.columns)
print(voters_df_2019.columns)
print(voters_df_2024.columns)

set(voters_df_2014.columns) == set(voters_df_2019.columns) == set(voters_df_2024.columns)
```

### In 2024 dataframe, we don't have STAES mentioned in STATE_NAME column. So, adding that information by looking at 2014 and 2019 dataframes


```python
state_name_dict = {}
for index, row in voters_df_2014.iterrows():
    state_name_dict[row['PC_NAME']] = row['STATE_NAME']
for index, row in voters_df_2019.iterrows():
    state_name_dict[row['PC_NAME']] = row['STATE_NAME'] 

voters_df_2024['STATE_NAME'] = voters_df_2024.apply(lambda row: None if row['PC_NAME'] not in state_name_dict.keys() else state_name_dict[row['PC_NAME']], axis=1)
```

### Adding year column to each dataframe


```python
voters_df_2014['YEAR'] = 2014
voters_df_2019['YEAR'] = 2019
voters_df_2024['YEAR'] = 2024
```

### Combining all dataframes into one

```python
voters_df = pd.concat([voters_df_2014, voters_df_2019, voters_df_2024], ignore_index=True)

voters_df.head()
```

### Filtering voters_df by list of constituencies

```python
list_of_10_costrituencies = ['Dhule', 'Jalgaon', 'Nagpur', 'Dindori', 'Nashik', 'Pune', 'Baramati', 'Shirur', 'Ahmednagar', 'Shirdi']

voters_df_10 = voters_df[voters_df['PC_NAME'].isin(list_of_10_costrituencies)]

voters_df_10.head(10)
```

### Data curation

#### Bifurcation of total voters as male, female and other is not available for 2019 and 2014 years.

```python
voters_df_10['MALE_ELECTORS'] = voters_df_10['EVM_MALE']/voters_df_10['VOTER_TURNOUT_MALE_PERCENT'] * 100
voters_df_10['FEMALE_ELECTORS'] = voters_df_10['EVM_FEMALE']/voters_df_10['VOTER_TURNOUT_FEMALE_PERCENT'] * 100
voters_df_10['OTHERS_ELECTORS'] = voters_df_10['EVM_OTHERS']/voters_df_10['VOTER_TURNOUT_OTHERS_PERCENT'] * 100
```

```python
voters_df_10.fillna(0, inplace=True)
```

#### Changing datatype of newly created columns to integer

```python
voters_df_10['MALE_ELECTORS'] = voters_df_10['MALE_ELECTORS'].astype(int)
voters_df_10['FEMALE_ELECTORS'] = voters_df_10['FEMALE_ELECTORS'].astype(int)
voters_df_10['OTHERS_ELECTORS'] = voters_df_10['OTHERS_ELECTORS'].astype(int)
```
